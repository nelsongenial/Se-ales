```python
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot
import numpy as np
from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display
```

```python
signal = thinkdsp.Chirp(start=220, end=11025/2)
wave1 = signal.make_wave(duration=2)
wave1.make_audio()
```

```python
signal2 = thinkdsp.BrownianNoise(amp = 44)
wave2 = signal2.make_wave(duration=2, framerate=11025)
wave2.make_audio()
```

```python
wave4 = wave1 + wave2
wave4.segment(start=0.0, duration=2).plot()
```

```python
wave4.make_audio()
```

```python
spectrums = wave4.make_spectrum()
spectrums.plot()
```
